Sometimes you want to perform a task repeatedly, without writing it out each time.  For example, here I will draw a comb with 30 teeth.  I don't want to write out a separate line for the tooth of each comb.  Instead I use a **for-loop**.

```
init()

jump_to(-35, 25)

for i in range(30): # the sequence of three commands underneath is repeated 30 times
    move(0, -32)
    move(0, 32)
    move(2, 0)

move(15, 0) # after the loop is finished, we draw the handle of the comb

draw(color='blue')
```

It can also be useful to have the loop perform a slightly different task each time.  We can do this by having the index *i* of the loop as an input to the instruction.  For example, here I will draw a feather.  I use the sine function so that the width of the feather increases, then decreases.

```
init()

jump_to(-10, -35)

for i in range(20):
  move(1, 3)
  w = np.sin(i/6.5) # the scale factor for the width of the feather at this step
  move(w*10, 0)
  move(-w*10, 0)
  move(-w*10, w*2)
  move(w*10, -w*2)

draw()
```

OK, time to do some dynamical systems.  From now on we'll always be working with a particular dynamical system, specified by *dx* and *dy*, which we'll draw in the background of all our pictures.

```
dx = lambda x, y: 0.1 * y
dy = lambda x, y: -0.1 * x

init()

move(1, 3)
move(4, 1)
move(7, 6)

draw_with_vector_background(dx, dy, color='red')
```
